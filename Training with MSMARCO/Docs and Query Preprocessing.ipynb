```python
#package
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def clean_data(text):
  lem = WordNetLemmatizer()
  lower_case = text.lower()
  tokenized = word_tokenize(lower_case)
  data_lemmatize=[lem.lemmatize(word,"v") for word in tokenized if not word in stopwords.words('english')]
  data_fixed = ' '.join(data_lemmatize)
  return data_fixed
```

```python
from tqdm.auto import tqdm
tqdm.pandas(desc="my bar!")
```

```python
training_corpus['cleaned'] = training_corpus['body'].progress_apply(lambda x: clean_data(x))
```

```python
testing_corpus['cleaned'] = testing_corpus['body'].progress_apply(lambda x: clean_data(x))
```

```python
training_queries['cleaned'] = training_queries['query'].progress_apply(lambda x: clean_data(x))
```

```python
testing_queries['cleaned'] = testing_queries['query'].progress_apply(lambda x: clean_data(x))
```
